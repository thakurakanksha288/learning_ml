```python
#basic linear model- logestic regression model
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model


```

```python
path=pd.read_csv("/content/canada_per_capita_income (1).csv")
path.head()
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('income')
plt.scatter(path.year, path['per capita income (US$)'], color="red", marker='x')
```

```python
reg=linear_model.LinearRegression()
reg.fit(path[['year']], path['per capita income (US$)'])
```

```python
reg.predict(pd.DataFrame([3300]))
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df=pd.read_csv("/content/homeprices.csv")
df
```

```python
#data preprocessin

import math
med_bedrooms=math.floor(df.bedrooms.median())
df.bedrooms = df.bedrooms.fillna(med_bedrooms)
df
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['area', 'bedrooms', 'age']], df.price)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict(pd.DataFrame([[3000, 3, 40]], columns=['area', 'bedrooms', 'age']))
reg.predict(pd.DataFrame([[2500, 4, 5]], columns=['area', 'bedrooms', 'age']))
```
