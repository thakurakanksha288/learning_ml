```python
import pandas as pd

df= pd.read_csv('/content/diabetes.csv')
df
```

```python
X=df.drop('Outcome',axis='columns')
y=df.Outcome
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
X_scaled[:3]
```

```python
from sklearn.model_selection import train_test_split


X_train,X_test,y_train,y_test=train_test_split(X_scaled,y,stratify=y,test_size=0.3,random_state=10)
```

```python
X_train.shape
```

```python
y_train.value_counts()
```

```python
187/350
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score

scores=cross_val_score(DecisionTreeClassifier(),X,y,cv=5)
scores
```

```python
scores.mean()
```

```python
from sklearn.ensemble import BaggingClassifier

bag_model=BaggingClassifier(
    estimator=DecisionTreeClassifier(),
    n_estimators=100,
    max_samples=0.8,
    oob_score=True,
    random_state=0
)

bag_model.fit(X_train,y_train)
bag_model.oob_score_
```

```python
bag_model.score(X_test,y_test)
```

```python
bag_model=BaggingClassifier(
    estimator=DecisionTreeClassifier(),
    n_estimators=100,
    max_samples=0.8,
    oob_score=True,
    random_state=0
)



score=cross_val_score(bag_model,X,y,cv=5).mean()

score
```

```python
from sklearn.ensemble import RandomForestClassifier

cross_val_score(RandomForestClassifier(n_estimators=50),X,y,cv=5).mean()
```
