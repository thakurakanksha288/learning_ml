```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
cols=["area","perimeter","compactness","length","width","assummetry","groove_length","class"]
df=pd.read_csv("seeds_dataset.txt",names=cols,sep='\s+', engine='python')
```

```python
df.head()
```

```python
for i in range(len(cols)-1):
  for j in range(i+1,len(cols)-1):
    x_label=cols[i]
    y_label=cols[j]
    sns.scatterplot(x=x_label,y=y_label,data=df,hue='class')
    plt.show()
```

```python
#clustering
from sklearn.cluster import KMeans
x='compactness'
y='assummetry'
x=df[[x,y]].values
```

```python
kmeans=KMeans(n_clusters=3).fit(x)
```

```python
clusters=kmeans.labels_
```

```python
clusters
```

```python
df["class"].values
```

```python
cluster_df=pd.DataFrame(np.hstack((x,clusters.reshape(-1,1))),columns=['perimeter','assummetry',"class"])
```

```python
#k means class
sns.scatterplot(x='perimeter',y='assummetry',hue="class",data=cluster_df)
plt.plot()
```

```python
#original classes
sns.scatterplot(x='perimeter',y='assummetry',hue="class",data=df)
plt.plot()
```

```python
#higher dimensions
X=df[cols[:-1]].values
```

```python
kmeans=KMeans(n_clusters=3).fit(X)
cluster_df=pd.DataFrame(np.hstack((X,kmeans.labels_.reshape(-1,1))),columns=df.columns)
```

```python
#k means class
sns.scatterplot(x='perimeter',y='assummetry',hue="class",data=cluster_df)
plt.plot()
```

```python
#original classes
sns.scatterplot(x='perimeter',y='assummetry',hue="class",data=df)
plt.plot()
```

```python
#pca
from sklearn.decomposition import PCA
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
transformed_x=pca.fit_transform(X)
```

```python
kmeans_pca_df=pd.DataFrame(np.hstack((transformed_x,kmeans.labels_.reshape(-1,1))),columns=["pca1","pca2","class"])
```

```python
from operator import truth
truth_pca_df=pd.DataFrame(np.hstack((transformed_x,df['class'].values.reshape(-1,1))),columns=["pca1","pca2","class"])
```

```python
#k means class
sns.scatterplot(x='pca1',y="pca2",hue="class",data=kmeans_pca_df)
plt.plot()
```

```python
#original classes
sns.scatterplot(x="pca1",y='pca2',hue="class",data=truth_pca_df)
plt.plot()
```
