```python
import pandas as pd
df=pd.read_csv('/content/salaries.csv')
df.head()
```

```python
imputs=df.drop('salary_more_then_100k',axis='columns')
target=df['salary_more_then_100k']
```

```python
#since ml algothim omly deal with numbers so we have to convert the data into numbers agaian

from sklearn.preprocessing import LabelEncoder
le_company=LabelEncoder()
le_job=LabelEncoder()
le_degree=LabelEncoder()
```

```python
imputs['company_n']=le_company.fit_transform(imputs['company'])
imputs['job_n']=le_job.fit_transform(imputs['job'])
imputs['degree_n']=le_degree.fit_transform(imputs['degree'])
imputs.head()
```

```python
imputs_n=imputs.drop(['company','job','degree'],axis='columns')
imputs_n
```

```python
from sklearn import tree
model=tree.DecisionTreeClassifier()
```

```python
model.fit(imputs_n,target)
```

```python
model.score(imputs_n,target)
```

```python
model.predict([[2,2,1]])
```
